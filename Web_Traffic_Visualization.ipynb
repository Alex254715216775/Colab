```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

months = ['07/2019', '08/2019', '09/2019', '10/2019', '11/2019']
searches = [50, 53, 59, 56, 62]
direct = [39, 47, 42, 51, 51]
social = [70, 80, 90, 87, 92]

bar_width = 0.25
x = np.arange(len(months))

plt.figure(figsize=(10, 6))
plt.bar(x - bar_width, searches, width=bar_width, label='Searches', color='deepskyblue')
plt.bar(x, direct, width=bar_width, label='Direct', color='lightcoral')
plt.bar(x + bar_width, social, width=bar_width, label='Social Media', color='orange')

plt.xlabel("months")
plt.ylabel("visitors\nin thousands", labelpad=15)
plt.title("Visitors by web traffic sources")
plt.xticks(x, months)
plt.legend()

for i in range(len(x)):
    plt.text(x[i] - bar_width, searches[i] + 1, str(searches[i]), ha='center')
    plt.text(x[i], direct[i] + 1, str(direct[i]), ha='center')
    plt.text(x[i] + bar_width, social[i] + 1, str(social[i]), ha='center')

plt.tight_layout()
plt.show()
```
